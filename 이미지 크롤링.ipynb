```python
from urllib import request
import urllib.request
from urllib.request import urlopen, urlretrieve
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import getpass
```

# 이미지 파일 만들어 놓기
- nav_img(네이버)/goo_img(구글)/ins_img(인스타) 폴더를 현재위치에 생성해야 거기에 저장됨

# 인스타 이미지 가져오기

```python
# 처음에 이거 코드 쓰기 전에 img 파일을 현재 디렉토리에 작성해두어야함

ID = input("인스타 아이디를 입력하세요 : ")
PW = getpass.getpass("인스타 비번을 입력하세요 : ")
search = input("검색할 태그를 입력하세요 :")


driver = webdriver.Chrome()
url = "https://www.instagram.com/"
driver.get(url)

time.sleep(1)

driver.find_element_by_name("username").send_keys(ID)
driver.find_element_by_name("password").send_keys(PW)
driver.find_element_by_class_name('y3zKF').click()

time.sleep(3)

element = driver.find_element_by_class_name("x3qfX")
element.send_keys(search)
time.sleep(1)
element.send_keys(Keys.ENTER)
element.send_keys(Keys.ENTER)

time.sleep(5)

body = driver.find_element_by_tag_name('body')
                        
for i in range(10):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)   

html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
img = soup.select(".FFVAD")    

imgUrl = []

for i in img:
    try:
        imgUrl.append(i.attrs['src'])
    except KeyError:
        imgUrl.append(i.attrs["data-src"])
        
n = 1        
for i in imgUrl:
    urlretrieve(i, "./ins_img/"+ search +str(n)+".jpg")
    n+=1
driver.close()
```

# 구글 이미지 가져오기

```python
search = input("검색할 태그를 입력하세요 :")
driver = webdriver.Chrome()
driver.get("https://www.google.co.kr/imghp?hl=ko&ogbl")
elem = driver.find_element_by_name("q")
#검색어 넣기
elem.send_keys(search)
elem.send_keys(Keys.RETURN)
 
body = driver.find_element_by_tag_name('body')

#스크롤 길이 조절
for i in range(3):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5) 
     
images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")

count = 200
for image in images:
    try: 
        image.click()
        time.sleep(1)
        imgUrl = driver.find_element_by_xpath('/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div/div/div[3]/div[2]/c-wiz/div/div[1]/div[1]/div/div[2]/a/img').get_attribute("src")
        urllib.request.urlretrieve(imgUrl,"./goo_img/" + search + str(count) + ".jpg")
        count = count + 1
    except:
        pass
```

# 네이버 이미지 가져오기

```python
search = input("검색할 태그를 입력하세요 :")
driver = webdriver.Chrome()
driver.get("https://search.naver.com/search.naver?where=image&sm=tab_jum&query=")
elem = driver.find_element_by_name("query")
#검색어 입력
elem.send_keys(search)
elem.send_keys(Keys.RETURN)

#스크롤 길이 조절
body = driver.find_element_by_tag_name('body')
                        
for i in range(10):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5) 
     
images = driver.find_elements_by_css_selector("._image._listImage")

count = 400
for image in images:
    try: 
        image.click()
        time.sleep(1)
        imgUrl = driver.find_element_by_xpath("//*[@id='main_pack']/section/div/div[2]/div/div[1]/div[1]/div[1]/div/div/div[1]/div[1]/img").get_attribute("src")
        urllib.request.urlretrieve(imgUrl,"./nav_img/" + search + str(count) + ".jpg")
        count = count + 1
    except:
        pass
```

# 다음 이미지 가져오기

```python
search = input("검색할 태그를 입력하세요 :")
driver = webdriver.Chrome()
driver.get("https://search.daum.net/search?w=img&nil_search=btn&DA=NTB&enc=utf8&q=")
elem = driver.find_element_by_name("q")
#검색어 넣기
elem.send_keys(search)
elem.send_keys(Keys.RETURN)
 
body = driver.find_element_by_tag_name('body')

#스크롤 길이 조절
for i in range(5):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5) 
     
images = driver.find_elements_by_css_selector(".thumb_img")

count = 600
for image in images:
    try: 
        image.click()
        time.sleep(1)
        imgUrl = driver.find_element_by_xpath("//*[@id='ins_img_viewer_big_img_org']").get_attribute("src")
        urllib.request.urlretrieve(imgUrl,"./daum_img/" + search + str(count) + ".jpg")
        count = count + 1
    except:
        pass
    
driver.close()
```

```python
def get(img):
    driver = webdriver.Chrome()
    URL = img
    driver.get(URL)
    imgUrl = driver.find_element_by_css_selector("body > img").get_attribute("src")
    urllib.request.urlretrieve(imgUrl, str(1) + ".jpg")
    return driver.close()
```
